```python
#While Loops
ts_length = 100
_values = []
i = 0
while i < ts_length: #this is another way to create a list of random normals of length 100
    e = np.random.randn() #but now we are using 'while' insted of 'for'
    _values.append(e)
    i = i + 1
plt.plot(_values)
plt.show()
#the statement i = i + 1 can be replaced by i += 1
```

```python
#User-Defined Functions
def generate_data(n):  #'def' creates a function in this case 'generate_data'
    _values = [] 
    for i in range(n):  
        e = np.random.randn()
        _values.append(e)
    return _values

data = generate_data(100)
plt.plot(data)
plt.show() #this shows the graph of the random Gaussian series
```

```python
#Conditions
def generate_data(n, generator_type): #create the function, creates a list of random numbers
    _values = []     #create the list
    for i in range(n):   #begin loop
        if generator_type == 'U':     #if distribution is uniform
            e = np.random.uniform(0, 1) #use uniform distribution
        else:
            e = np.random.randn()  #if not specified use Gaussian distribution
        _values.append(e)  #append just left joins the new number
    return _values #return, specifies for the function to output the created list
```

```python
data = generate_data(100, 'U') 
plt.plot(data)
plt.show()
```

```python
data = generate_data(100,'asdf')
plt.plot(data)
plt.show()
```

```python
#generate random list of numbers
def generate_data(n, generator_type): 
    _values = []
    for i in range(n):
        e = generator_type()
        _values.append(e)
    return _values
```

```python
#for random normal distribution
data = generate_data(100, np.random.randn)
plt.plot(data)
plt.show()
```

```python
#for random uniform distribution
data = generate_data(200, np.random.uniform)
plt.plot(data)
plt.show()
```

```python
#Exercise 1
def factorial(x):
    aux_values = [i+1 for i in range(x)]
    j = 1
    for i in aux_values:
        j = j*i
    return(j)
```

```python
factorial(5)
```

```python
#Their solution for Exercise 1
def factorial(n):
    k = 1
    for i in range(n):
        k = k * (i + 1)
    return k

factorial(4)
```

```python
#Exercise 2
from numpy.random import uniform
def binomial_rv(n,p):
    aux1 = 0
    for ii in range(n):
        e = uniform()
        if e < p:
            aux1 += 1
    return(aux1)

binomial_rv(10,.9)
```

```python
#Their solution for Exercise 2
from numpy.random import uniform

def binomial_rv(n, p):
    count = 0
    for i in range(n):
        U = uniform()
        if U < p:
            count = count + 1    # Or count += 1
    return count

binomial_rv(10, 0.5)
```

```python
#Exercise 3
import numpy as np
n = 1000
aux1 = 0
for ii in range(n):
    x = np.random.uniform()-.5
    y = np.random.uniform()-.5
    r = (x**2 + y**2)**.5
    
    if r < .5:
        aux1 += 1

prop1 = aux1/n
found_pi = prop1*4
print(found_pi)
```

```python
#Their solution for Exercise 3
n = 100000

count = 0
for i in range(n):
    u, v = np.random.uniform(), np.random.uniform()
    d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2)
    if d < 0.5:
        count += 1

area_estimate = count / n

print(area_estimate * 4)  # dividing by radius**2
```
